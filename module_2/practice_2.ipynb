Привет! В задаче нужно найти математическое ожидание и дисперсию расстояния от случайной точки \( A \) до центра круга радиуса \( R = 3 \).

### Шаг 1: Найдём математическое ожидание расстояния

Пусть \( \xi \) — это расстояние от случайной точки \( A \) до центра круга. Так как точки распределены равномерно по кругу, вероятность того, что точка окажется на расстоянии \( r \) от центра, пропорциональна длине окружности радиуса \( r \). 

Функция плотности вероятности для расстояния \( r \) будет пропорциональна \( r \) (т.к. длина окружности радиуса \( r \) равна \( 2\pi r \)):

\[
$$f(r) = \frac{2r}{R^2}, \quad 0 \leq r \leq R
\]
где \( R = 3 \).$$

Математическое ожидание \( \mathbb{E}[\xi] \) вычисляется как интеграл:

\[
\mathbb{E}[\xi] = \int_0^R r f(r) \, dr = \int_0^R r \cdot \frac{2r}{R^2} \, dr = \frac{2}{R^2} \int_0^R r^2 \, dr
\]

Вычислим интеграл:

\[
\int_0^R r^2 \, dr = \frac{R^3}{3}
\]

Таким образом,

\[
\mathbb{E}[\xi] = \frac{2}{R^2} \cdot \frac{R^3}{3} = \frac{2R}{3}
\]

Подставляем \( R = 3 \):

\[
\mathbb{E}[\xi] = \frac{2 \cdot 3}{3} = 2
\]

### Шаг 2: Найдём дисперсию

Дисперсия \( \text{Var}(\xi) \) вычисляется по формуле:

\[
\text{Var}(\xi) = \mathbb{E}[\xi^2] - (\mathbb{E}[\xi])^2
\]

Сначала вычислим \( \mathbb{E}[\xi^2] \):

\[
\mathbb{E}[\xi^2] = \int_0^R r^2 f(r) \, dr = \int_0^R r^2 \cdot \frac{2r}{R^2} \, dr = \frac{2}{R^2} \int_0^R r^3 \, dr
\]

Вычислим интеграл:

\[
\int_0^R r^3 \, dr = \frac{R^4}{4}
\]

Тогда:

\[
\mathbb{E}[\xi^2] = \frac{2}{R^2} \cdot \frac{R^4}{4} = \frac{R^2}{2}
\]

Подставляем \( R = 3 \):

\[
\mathbb{E}[\xi^2] = \frac{3^2}{2} = \frac{9}{2}
\]

Теперь можем найти дисперсию:

\[
\text{Var}(\xi) = \mathbb{E}[\xi^2] - (\mathbb{E}[\xi])^2 = \frac{9}{2} - 2^2 = \frac{9}{2} - 4 = \frac{1}{2}
\]

### Ответы:
1. Математическое ожидание \( \mathbb{E}[\xi] = 2 \)
2. Дисперсия \( \text{Var}(\xi) = \frac{1}{2} \)

### Ключевая концепция:

### Пояснение концепции:
Математическое ожидание случайной величины показывает среднее значение, которое она принимает. В случае равномерного распределения внутри круга это значение находится с учетом плотности вероятности, зависящей от радиуса. Дисперсия показывает, насколько значения случайной величины отклоняются от её среднего значения, что также рассчитывается через интеграл.
